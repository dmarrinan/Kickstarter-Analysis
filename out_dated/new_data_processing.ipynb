```python
cat_list = []
cat_id_list = []

backer_counts = []
blurbs = []
campaign_lengths = []
cat_ids = []
cat_names = []
parent_cat_ids = []
countries = []
creators = []
currencies = []
deadlines = []
goals = []
ids = []
start_dates = []
states = []
slugs = []
spotlights = []
results = []
conversion_rates = []
pledged_amounts = []

for i in range(0,14):
    if i < 10:
        file_location = f'raw_data/Kickstarter00{i}.csv'
    else:
        file_location = f'raw_data/Kickstarter0{i}.csv'
    sheet_df = pd.read_csv(file_location)

    for cell in sheet_df['projects']:
        campaign_list = json.loads(cell)

        for project in campaign_list:
            backer_count = project['backers_count']
            blurb = project['blurb']
            cat_id = project['category']['id']
            cat_name = project['category']['name']
            try:
                parent_cat_id = project['category']['parent_id']
            except KeyError:
                parent_cat_id = '0'
            country = project['country']
            creator = project['creator']['urls']['web']['user']
            currency = project['currency']
            start_date_ts = project['launched_at']
            deadline_ts = project['deadline']
            campaign_length = round((deadline_ts - start_date_ts)/86400,2)
            start_date = dt.datetime.fromtimestamp(int(start_date_ts)).strftime('%Y-%m-%d %H:%M:%S')
            deadline = dt.datetime.fromtimestamp(int(deadline_ts)).strftime('%Y-%m-%d %H:%M:%S')
            goal = project['goal']
            id = project['id']
            try:
                state = project['location']['state']
            except KeyError:
                state = 'na'        
            slug = project['slug']
            spotlight = project['spotlight']
            result = project['state']
            conversion_rate = project['static_usd_rate']
            pledged_amount = project['pledged']

            backer_counts.append(backer_count)
            blurbs.append(blurb)
            cat_ids.append(cat_id)
            cat_names.append(cat_name)
            campaign_lengths.append(campaign_length)
            parent_cat_ids.append(parent_cat_id)
            countries.append(country)
            creators.append(creator)
            currencies.append(currency)
            deadlines.append(deadline)
            goals.append(goal)
            ids.append(id)
            start_dates.append(start_date)
            states.append(state)
            slugs.append(slug)
            spotlights.append(spotlight)
            results.append(result)
            conversion_rates.append(conversion_rate)
            pledged_amounts.append(pledged_amount)

            if cat_id in cat_id_list:
                cat_id_list = cat_id_list
            else:
                cat_id_list.append(cat_id)
                cat_list.append(cat_name)

project_df = pd.DataFrame({
    "Backers":backer_counts,
    "Blurb":blurbs,
    "Campaign Length (Days)":campaign_lengths,
    "Category ID":cat_ids,
    "Category Name":cat_names,
    "Parent Category ID":parent_cat_ids,
    "Country":countries,
    "Creator":creators,
    "Currency":currencies,
    "Deadline":deadlines,
    "Goal":goals,
    "Project ID":ids,
    "Start Date":start_dates,
    "State":states,
    "Slug":slugs,
    "Spotlight":spotlights,
    "Result":results,
    "Conversion Rate":conversion_rates,
    "Amount Pledged":pledged_amounts,
})
```
