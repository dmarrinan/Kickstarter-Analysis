```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#load data

full_df = pd.read_csv('raw_data/kickstarter_projects.csv', low_memory=False)
full_df.head()
```

```python
#remove ongoing campaigns and misformatted rows

success_df = full_df[full_df.state == 'successful']
failure_df = full_df[full_df.state == 'failed']
canceled_df = full_df[full_df.state == 'canceled']
dfs = [success_df, failure_df, canceled_df]
kickstarter_df = pd.concat(dfs)
kickstarter_df.head()
```

```python
#extract only features of interest

features_of_interest = pd.DataFrame({
    "kickstarter_id":kickstarter_df["kickstarter_id"],
    "name":kickstarter_df["name"],
    "slug":kickstarter_df["slug"],
    "blurb":kickstarter_df["blurb"],
    "main_category":kickstarter_df["main_category"],
    "sub_category":kickstarter_df["sub_category"],
    "launch_date":kickstarter_df["launched_at"],
    "deadline_date":kickstarter_df["deadline"],
    "result":kickstarter_df["state"],
    "goal":kickstarter_df["goal"],
    "pledged":kickstarter_df["pledged"],
    "pct_funded":kickstarter_df["percentage_funded"],
    "backer_count":kickstarter_df["backers_count"],
    "currency":kickstarter_df["currency"]
})

print(len(features_of_interest.index))
features_of_interest.head()
```

```python
blurbs = features_of_interest["blurb"]
blurb_lengths = []
blurb_sentiment_compound = []
blurb_sentiment_positive = []
blurb_sentiment_negative = []
blurb_sentiment_neutral = []
for blurb in blurbs:
    try:
        blurb_length = len(blurb)
        polarity_results = analyzer.polarity_scores(blurb)
        sentiment_compound = polarity_results['compound']
        sentiment_positive = polarity_results['pos']
        sentiment_negative = polarity_results['neg']
        sentiment_neutral = polarity_results['neu']
    except TypeError:
        blurb_length = -1
        sentiment_compound = 0
        sentiment_positive = 0
        sentiment_negative = 0
        sentiment_neutral = 0
    blurb_sentiment_compound.append(sentiment_compound)
    blurb_sentiment_positive.append(sentiment_positive)
    blurb_sentiment_negative.append(sentiment_negative)
    blurb_sentiment_neutral.append(sentiment_neutral)
    blurb_lengths.append(blurb_length)

features_of_interest['blurb_length'] = blurb_lengths
features_of_interest['blurb_sentiment_compound'] = blurb_sentiment_compound
features_of_interest['blurb_sentiment_positive'] = blurb_sentiment_positive
features_of_interest['blurb_sentiment_negative'] = blurb_sentiment_negative
features_of_interest['blurb_sentiment_neutral'] = blurb_sentiment_neutral

features_of_interest = features_of_interest.query('blurb_length != -1')

print(len(features_of_interest.index))
features_of_interest.head()
```

```python
#remove time from date columns
features_of_interest['deadline_date'] = features_of_interest['deadline_date'].str.split(' ').str[0]
features_of_interest['launch_date'] = features_of_interest['launch_date'].str.split(' ').str[0]
features_of_interest.head()
```

```python
#create datetime columns to represent dates
features_of_interest['deadline_date_dtobject'] = pd.to_datetime(features_of_interest.loc[:,'deadline_date'])
features_of_interest['launch_date_dtobject'] = pd.to_datetime(features_of_interest.loc[:,'launch_date'])
features_of_interest.head()
```

```python
features_of_interest['num_days'] = (features_of_interest['deadline_date_dtobject'] - features_of_interest['launch_date_dtobject']) / np.timedelta64(1, 'D')
features_of_interest.head()
```

```python
print(f'Earliest Launch Date: {min(features_of_interest["launch_date_dtobject"])}')
print(f'Latest Launch Date: {max(features_of_interest["launch_date_dtobject"])}')
print(f'Earliest Deadline Date: {min(features_of_interest["deadline_date_dtobject"])}')
print(f'Earliest Deadline Date: {max(features_of_interest["deadline_date_dtobject"])}')
```
